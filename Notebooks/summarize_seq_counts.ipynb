Read htseq-count "*.count" result file and summarize special counters

```python

rez_table = pd.DataFrame(columns = ['file','__no_feature','__ambiguous','__too_low_aQual','__not_aligned','__alignment_not_unique'])

# Create new data frame with sequence count from HTSeq result files
for i in files:
    try:
        # read csv from file
        rgi_rez = pd.read_csv(i, sep="	", header=None)
                
        # extract sample name from file path
        start = i.find(pwd) + len(pwd)
        end = i.find('.csv', start)
                
        # write result data to combined table
        rez_table.loc[len(rez_table)] = [i[start:end], rgi_rez.iloc[-5][1], rgi_rez.iloc[-4][1], rgi_rez.iloc[-3][1], rgi_rez.iloc[-2][1], rgi_rez.iloc[-1][1] ]
        print(rez_table)
    except:
        print("FILE EMPTY:", i,)

rez_table
```

## Calculate special counter Total and avarage values

```python
feature = 0 
ambiogious = 0 
low_qualit = 0
not_alligned = 0
al_not_unique = 0

size = len(rez_table)

for i in range(len(rez_table)):
    feature += rez_table.iloc[i].loc["__no_feature"]
    ambiogious += rez_table.iloc[i].loc["__ambiguous"]
    low_qualit += rez_table.iloc[i].loc["__too_low_aQual"]
    not_alligned += rez_table.iloc[i].loc["__not_aligned"]
    al_not_unique += rez_table.iloc[i].loc["__alignment_not_unique"]

rez_table.loc[len(rez_table)] = ["Total", feature, ambiogious, low_qualit, not_alligned, al_not_unique ]

```

```python
rez_table.loc[len(rez_table)] = ["Avarage", feature/size, ambiogious/size, low_qualit/size, not_alligned/size, al_not_unique/size ]
```

## write results to csv file

```python
rez_table.to_csv('special_counters.csv',index=False)

```

# Count AMR gene variants

```python

gene_table = pd.DataFrame()

for file in files:
    try:
        # extract sample name from file path
        start = file.find('htseq_filtered/') + 15
        end = file.find('.csv', start)

        # add sample column
        gene_table[file[start:end]] = 0

        rgi_rez = pd.read_csv(file, sep="	", header=None)

        # Add gene counts 
        for i in range(len(rgi_rez)-5):

            # if gene name is not in the table, add new row.
            if((gene_table.index !=  rgi_rez.iloc[i][0]).all()):
                gene_table.loc[rgi_rez.iloc[i][0]] = 0
            
            # Add gene count to table 
            gene_table[file[start:end]][rgi_rez.iloc[i][0]] += rgi_rez.iloc[i][1]

        print(gene_table)
    except:
        print("FILE EMPTY:", file)
```

```python
# Calculate total and mean values
rez2 = gene_table.sum(axis=1)
rez3 = gene_table.mean(axis=1)

## Add total count of genes
gene_table["Total"] = rez2

## Calculate row sum
rez1 = gene_table.sum(axis=0)
gene_table.loc["Total"] = rez1

## Add mean gene values in samples
gene_table["Mean"] = rez3

print("Total :")
print(gene_table["Total"])
print("Mean :")
print(gene_table["Mean"])
```

# Parse DEEPARG result files

Open result file and print content example

Combine all result files into table

```python
import pandas as pd
import glob

# Combine ARG results.
# Create new table with ARG types and counts
# Form ARG collumn in deeparg_rez take unique arge name
# And combine counts of same ARGs

# Create new table with ARG types and counts
arg_table = pd.DataFrame(columns = ['ARG'])

dir_files = glob.glob("../../NAS/bioinfo/edgars.liepa/Antibiotic_resistance/ww_publication/ARG_deepARG-ss/*.clean.deeparg.mapping.ARG")

for file in dir_files:
    # extract sample name from file path
    deeparg_rez = pd.read_csv(file, sep="	", header=0)
    
    start = len("../../NAS/bioinfo/edgars.liepa/Antibiotic_resistance/ww_publication/ARG_deepARG-ss/")
    end = file.find('.clean', start)

    # add sample column
    arg_table[file[start:end]] = 0
    
    unique_values = deeparg_rez['#ARG'].unique()

    for unique_value in unique_values:  
        if ((arg_table['ARG'] != unique_value).all()):
            arg_table.loc[len(arg_table)] = [unique_value] + [0]*(len(arg_table.columns) - 1)

    # # And combine counts of same ARGs
    for i in range(len(arg_table)):
        arg_table[file[start:end]][i] = deeparg_rez[deeparg_rez['#ARG'] == arg_table['ARG'][i]]['counts'].sum()

# Sort table by count
# arg_table = arg_table.sort_values(by=['Count'], ascending=False)

arg_table
    
```
